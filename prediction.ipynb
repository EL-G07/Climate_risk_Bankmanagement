```python
#fig=px.bar(data, x='Institution', y='Climate_risk_tier')
#app.layout=html.Div([html.H1("Institution Dashbord"),
#                     dcc.Graph(figure=fig)])

app.layout = html.Div([
    # ===== DARK BACKGROUND FOR ENTIRE DASHBOARD =====
    html.Div(style={
        'backgroundColor': '#0d1117',
        'minHeight': '100vh',
        'padding': '30px',
        'position': 'fixed',
        'top': 0,
        'left': 0,
        'right': 0,
        'bottom': 0,
        'overflowY': 'auto'
    }, children=[
        
        # ===== BACK BUTTON - TOP RIGHT CORNER =====
        html.Div([
            html.A(
                html.Button(
                    "← Back to Home",
                    id="back-home-btn",
                    style={
                        'backgroundColor': '#00c853',
                        'color': '#0d1117',
                        'padding': '10px 25px',
                        'border': 'none',
                        'borderRadius': '8px',
                        'fontSize': '16px',
                        'fontWeight': 'bold',
                        'cursor': 'pointer',
                        'transition': 'all 0.3s ease',
                        'boxShadow': '0 0 15px #00c85344'
                    }
                ),
                href="http://127.0.0.1:8050",  # Change this to your home page URL
                target="_self",
                style={'textDecoration': 'none'}
            )
        ], style={
            'textAlign': 'right',
            'marginBottom': '20px',
            'position': 'sticky',
            'top': 0,
            'zIndex': 1000,
            'backgroundColor': '#0d1117',
            'padding': '10px 0'
        }),
        
        # ===== TABLE AT THE VERY TOP =====
        html.Div([
            html.H3("Bank Summary Table", style={
                'color': '#00c853', 
                'marginBottom': '20px',
                'fontWeight': 'bold',
                'letterSpacing': '1px'
            }),
            html.Div(id="summary-table", style={
                'overflowX': 'auto',
                'padding': '20px',
                'backgroundColor': '#161b22',
                'borderRadius': '10px',
                'border': f'1px solid #00c853',
                'boxShadow': f'0 0 20px #00c85333',
                'marginBottom': '30px'
            })
        ], style={'marginBottom': '30px'}),
        
        # ===== BANK DASHBOARD SECTION =====
        html.Div([
            html.Div([
                html.H3(" Bank Risk Dashboard", style={
                    'color': '#00c853', 
                    'marginBottom': '20px',
                    'fontWeight': 'bold'
                }),
                dcc.Graph(id="stacked-bar-chart", config={'displayModeBar': True})
            ], style={
                'padding': '20px',
                'backgroundColor': '#161b22',
                'borderRadius': '10px',
                'border': f'1px solid #00c853',
                'boxShadow': f'0 0 20px #00c85333',
                'marginBottom': '30px'
            }),
            
            # Bottom section with bank details and speedometer
            html.Div([
                # Left column - Bank selector and summary
                html.Div([
                    html.H3("Bank Details", style={
                        'color': '#00c853', 
                        'marginBottom': '20px',
                        'fontWeight': 'bold'
                    }),
                    html.Label("Select Bank:", style={
                        'fontWeight': 'bold', 
                        'display': 'block', 
                        'marginBottom': '5px',
                        'color': '#f0f6fc'
                    }),
                    dcc.Dropdown(
                        id="bank-selector",
                        options=[{'label': bank, 'value': bank} for bank in df['Institution'].unique()],
                        value=df['Institution'].iloc[0],
                        style={'marginBottom': '20px', 'backgroundColor': '#161b22', 'color': '#f0f6fc'}
                    ),
                    html.Div(id="bank-summary", style={
                        'padding': '15px',
                        'backgroundColor': '#0d1117',
                        'borderRadius': '5px',
                        'marginBottom': '15px',
                        'color': '#f0f6fc',
                        'border': '1px solid #00c85333'
                    }),
                    html.Div(id="investment-details", style={
                        'padding': '15px',
                        'backgroundColor': '#0d1117',
                        'borderRadius': '5px',
                        'color': '#f0f6fc',
                        'border': '1px solid #00c85333'
                    })
                ], style={
                    'width': '45%',
                    'display': 'inline-block',
                    'verticalAlign': 'top',
                    'padding': '20px',
                    'backgroundColor': '#161b22',
                    'borderRadius': '10px',
                    'border': f'1px solid #00c853',
                    'boxShadow': f'0 0 20px #00c85333',
                    'marginRight': '5%'
                }),
                html.Div([
                    html.H3("Risk Assessment Speedometer", style={
                        'color': '#00c853', 
                        'textAlign': 'center', 
                        'marginBottom': '20px',
                        'fontWeight': 'bold'
                    }),
                    dcc.Graph(id="gauge", config={'displayModeBar': False})
                ], style={
                    'width': '45%',
                    'display': 'inline-block',
                    'verticalAlign': 'top',
                    'padding': '20px',
                    'backgroundColor': '#161b22',
                    'borderRadius': '10px',
                    'border': f'1px solid #00c853',
                    'boxShadow': f'0 0 20px #00c85333',
                    'boxSizing': 'border-box'
                })
            ], style={'marginBottom': '30px'}),
            
            # ===== YOUR INVESTMENT PORTFOLIO - MOVED HERE =====
            html.Div([
                html.H3("Your Investment Portfolio", style={
                    'color': '#00c853', 
                    'marginBottom': '20px',
                    'fontWeight': 'bold'
                }),
                html.Div([
                    # Input fields for each sector
                    html.Div([
                        html.Div([
                            html.Label("Agriculture Investment (%):", style={
                                'fontWeight': 'bold', 
                                'display': 'block', 
                                'marginBottom': '5px',
                                'color': '#f0f6fc'
                            }),
                            dcc.Input(
                                id="user-agri",
                                type="number",
                                placeholder="Enter %",
                                value=25,
                                min=0,
                                max=100,
                                style={'width': '100%', 'padding': '8px', 'borderRadius': '5px', 'border': '1px solid #00c853', 'backgroundColor': '#161b22', 'color': '#f0f6fc'}
                            ),
                        ], style={'width': '22%', 'display': 'inline-block', 'marginRight': '2%'}),
                        
                        html.Div([
                            html.Label("Coastal Investment (%):", style={
                                'fontWeight': 'bold', 
                                'display': 'block', 
                                'marginBottom': '5px',
                                'color': '#f0f6fc'
                            }),
                            dcc.Input(
                                id="user-coastal",
                                type="number",
                                placeholder="Enter %",
                                value=25,
                                min=0,
                                max=100,
                                style={'width': '100%', 'padding': '8px', 'borderRadius': '5px', 'border': '1px solid #00c853', 'backgroundColor': '#161b22', 'color': '#f0f6fc'}
                            ),
                        ], style={'width': '22%', 'display': 'inline-block', 'marginRight': '2%'}),
                        
                        html.Div([
                            html.Label("Flood Zone Investment (%):", style={
                                'fontWeight': 'bold', 
                                'display': 'block', 
                                'marginBottom': '5px',
                                'color': '#f0f6fc'
                            }),
                            dcc.Input(
                                id="user-flood",
                                type="number",
                                placeholder="Enter %",
                                value=25,
                                min=0,
                                max=100,
                                style={'width': '100%', 'padding': '8px', 'borderRadius': '5px', 'border': '1px solid #00c853', 'backgroundColor': '#161b22', 'color': '#f0f6fc'}
                            ),
                        ], style={'width': '22%', 'display': 'inline-block', 'marginRight': '2%'}),
                        
                        html.Div([
                            html.Label("Energy Investment (%):", style={
                                'fontWeight': 'bold', 
                                'display': 'block', 
                                'marginBottom': '5px',
                                'color': '#f0f6fc'
                            }),
                            dcc.Input(
                                id="user-energy",
                                type="number",
                                placeholder="Enter %",
                                value=25,
                                min=0,
                                max=100,
                                style={'width': '100%', 'padding': '8px', 'borderRadius': '5px', 'border': '1px solid #00c853', 'backgroundColor': '#161b22', 'color': '#f0f6fc'}
                            ),
                        ], style={'width': '22%', 'display': 'inline-block'}),
                        
                    ], style={'marginBottom': '15px'}),
                    
                    html.Div([
                        html.Button(
                            "Calculate My Risk",
                            id="calculate-button",
                            n_clicks=0,
                            style={
                                'backgroundColor': '#00c853',
                                'color': '#0d1117',
                                'padding': '10px 30px',
                                'border': 'none',
                                'borderRadius': '5px',
                                'fontSize': '16px',
                                'cursor': 'pointer',
                                'fontWeight': 'bold',
                                'transition': 'all 0.3s ease'
                            }
                        ),
                        html.Button(
                            " Reset",
                            id="reset-button",
                            n_clicks=0,
                            style={
                                'backgroundColor': '#95a5a6',
                                'color': 'white',
                                'padding': '10px 30px',
                                'border': 'none',
                                'borderRadius': '5px',
                                'fontSize': '16px',
                                'cursor': 'pointer',
                                'fontWeight': 'bold',
                                'marginLeft': '10px'
                            }
                        ),
                    ], style={'textAlign': 'center', 'marginBottom': '20px'}),
                    
                    # User results section
                    html.Div(id="user-results", style={
                        'padding': '20px',
                        'backgroundColor': '#161b22',
                        'borderRadius': '10px',
                        'marginBottom': '30px',
                        'border': '2px dashed #00c853',
                        'color': '#f0f6fc'
                    }),
                ], style={
                    'padding': '20px',
                    'backgroundColor': '#161b22',
                    'borderRadius': '10px',
                    'border': f'1px solid #00c853',
                    'boxShadow': f'0 0 20px #00c85333',
                    'marginBottom': '30px'
                }),
            ]),
            
            # ===== LEGEND AND INSTRUCTIONS =====
            html.Div([
                html.H4("Legend & Instructions:", style={
                    'color': '#00c853',
                    'fontWeight': 'bold'
                }),
                html.Div([
                    html.Span("Low Risk (0-33%)", style={'marginRight': '20px', 'color': '#2ECC71', 'fontWeight': 'bold'}),
                    html.Span("Medium Risk (33-66%)", style={'marginRight': '20px', 'color': '#F1C40F', 'fontWeight': 'bold'}),
                    html.Span("High Risk (66-100%)", style={'marginRight': '20px', 'color': '#E74C3C', 'fontWeight': 'bold'}),
                ], style={'marginBottom': '10px'}),
                html.P("Enter your investment percentages above and click 'Calculate My Risk' to see your personal risk assessment.", 
                       style={'color': '#8b949e'}),
                html.P("Use the dropdown below to view detailed investment breakdown and climate risk assessment for each bank.", 
                       style={'color': '#8b949e'})
            ], style={
                'padding': '20px',
                'backgroundColor': '#161b22',
                'borderRadius': '10px',
                'border': f'1px solid #00c853',
                'boxShadow': f'0 0 20px #00c85333',
                'marginTop': '20px'
            })
        ])
    ])
])  # ← This closes the main Div

# ===== FUNCTION TO CREATE SPEEDOMETER =====
def create_speedometer(risk_score, risk_tier, color, height=300):
    """Create a speedometer gauge for risk assessment"""
    return go.Figure(go.Indicator(
        mode="gauge+number",
        value=risk_score * 100,  # Convert to percentage
        title={
            "text": f"<b>{risk_tier} Risk</b>",
            "font": {"size": 20, "color": color}
        },
        number={
            "suffix": "%",
            "valueformat": ".1f",
            "font": {"size": 35, "color": color}
        },
        gauge={
            "axis": {
                "range": [0, 100],
                "tickmode": "array",
                "tickvals": [0, 25, 50, 75, 100],
                "ticktext": ["0", "25", "50", "75", "100"],
                "tickfont": {"size": 12}
            },
            "bar": {
                "color": color,
                "thickness": 0.2
            },
            "bgcolor": "white",
            "borderwidth": 2,
            "bordercolor": "#2c3e50",
            "steps": [
                {"range": [0, 33], "color": "blue"},
                {"range": [33, 66], "color": "yellow"},
                {"range": [66, 100], "color": "red"},
            ],
            "threshold": {
                "line": {"color": "black", "width": 4},
                "thickness": 0.6,
                "value": risk_score * 100
            }
        }
    )).update_layout(
        height=height,
        margin=dict(t=50, b=30, l=30, r=30),
        paper_bgcolor='rgba(0,0,0,0)',
        plot_bgcolor='rgba(0,0,0,0)'
    )

# ===== USER CALLBACK =====
@app.callback(
    Output("user-results", "children"),
    [Input("calculate-button", "n_clicks"),
     Input("reset-button", "n_clicks")],
    [State("user-agri", "value"),
     State("user-coastal", "value"),
     State("user-flood", "value"),
     State("user-energy", "value")]
)
def calculate_user_risk(calc_clicks, reset_clicks, agri, coastal, flood, energy):
    ctx = callback_context
    if not ctx.triggered:
        return html.Div([
            html.P(" Enter your investment percentages above and click 'Calculate My Risk' to see your results.", 
                   style={'fontSize': '16px', 'color': '#8b949e'})
        ])
    
    button_id = ctx.triggered[0]['prop_id'].split('.')[0]
    
    if button_id == "reset-button":
        return html.Div([
            html.P(" Reset complete! Enter new values and click 'Calculate My Risk'.", 
                   style={'fontSize': '16px', 'color': '#8b949e'})
        ])
    
    if button_id == "calculate-button":
        # Validate inputs
        inputs = [agri, coastal, flood, energy]
        if any(v is None or v == '' for v in inputs):
            return html.Div([
                html.P(" Please enter valid numbers for all investment fields.", 
                       style={'fontSize': '16px', 'color': '#E74C3C', 'fontWeight': 'bold'})
            ])
        
        # Convert to float
        agri = float(agri)
        coastal = float(coastal)
        flood = float(flood)
        energy = float(energy)
        
        # Check total
        total = agri + coastal + flood + energy
        if total > 100:
            return html.Div([
                html.P(f" Total investment must equal 100%. Current total: {total}%", 
                       style={'fontSize': '16px', 'color': '#E74C3C', 'fontWeight': 'bold'})
            ])
        
        # Calculate risk score using fr weights
        risk_score = (fr.agri_w * agri + 
                      fr.cst_w * coastal + 
                      fr.fld_w * flood + 
                      fr.en_w * energy) / (fr.total_w * 100)
        
        # Determine risk tier
        if risk_score < 0.33:
            risk_tier = "Low"
            color = "#2ECC71"
        elif risk_score < 0.66:
            risk_tier = "Medium"
            color = "#F1C40F"
        else:
            risk_tier = "High"
            color = "#E74C3C"
        
        # Create speedometer
        user_gauge = create_speedometer(risk_score, risk_tier, color, height=300)
        
        # Return user results
        return html.Div([
            html.H4("Your Personal Risk Assessment", style={'color': '#00c853', 'marginBottom': '15px'}),
            html.Div([
                # Left side - Summary
                html.Div([
                    html.Strong("Your Portfolio Breakdown:", style={'display': 'block', 'fontSize': '16px', 'marginBottom': '10px', 'color': '#f0f6fc'}),
                    html.Div([
                        html.Span(f"Agriculture: {agri}%", style={'display': 'block', 'marginBottom': '5px', 'color': '#f0f6fc'}),
                        html.Span(f" Coastal: {coastal}%", style={'display': 'block', 'marginBottom': '5px', 'color': '#f0f6fc'}),
                        html.Span(f" Flood Zone: {flood}%", style={'display': 'block', 'marginBottom': '5px', 'color': '#f0f6fc'}),
                        html.Span(f" Energy: {energy}%", style={'display': 'block', 'marginBottom': '5px', 'color': '#f0f6fc'}),
                    ]),
                    html.Hr(style={'borderColor': '#00c85333'}),
                    html.Div([
                        html.Span(f" Risk Score: {(risk_score*100):.1f}%", 
                                 style={'fontSize': '18px', 'fontWeight': 'bold', 'display': 'block', 'color': '#f0f6fc'}),
                        html.Span(f" Risk Tier: ", style={'display': 'inline', 'fontSize': '16px', 'color': '#f0f6fc'}),
                        html.Span(risk_tier, style={
                            'color': color,
                            'fontWeight': 'bold',
                            'fontSize': '20px',
                            'display': 'inline'
                        }),
                    ])
                ], style={
                    'width': '45%',
                    'display': 'inline-block',
                    'verticalAlign': 'top',
                    'padding': '15px',
                    'backgroundColor': '#0d1117',
                    'borderRadius': '8px',
                    'border': '1px solid #00c85333'
                }),
                # Right side - Speedometer
                html.Div([
                    dcc.Graph(
                        id='user-gauge-display',
                        figure=user_gauge,
                        config={'displayModeBar': False}
                    )
                ], style={
                    'width': '50%',
                    'display': 'inline-block',
                    'verticalAlign': 'top',
                    'padding': '10px'
                })
            ], style={'display': 'flex', 'flexWrap': 'wrap'}),
            # Comparison with banks
            html.Div([
                html.H5(" How You Compare to Banks:", style={'marginTop': '20px', 'color': '#00c853'}),
                html.Div([
                    html.Span(f"Your Risk: {risk_tier} ({ (risk_score*100):.1f}%)", 
                             style={'fontWeight': 'bold', 'color': color, 'marginRight': '20px'}),
                    html.Span(" Bank Average Risk: ", style={'fontWeight': 'bold', 'color': '#f0f6fc'}),
                    html.Span(f"{df['Climate_risk_tier'].value_counts().idxmax()} Risk", 
                             style={'color': risk_colors.get(df['Climate_risk_tier'].value_counts().idxmax(), 'black')})
                ])
            ], style={'marginTop': '20px', 'padding': '15px', 'backgroundColor': '#0d1117', 'borderRadius': '8px', 'border': '1px solid #00c85333'})
        ])
    
    return html.Div()

# ===== BANK DASHBOARD CALLBACK =====
@app.callback(
    [Output("stacked-bar-chart", "figure"),
     Output("gauge", "figure"),
     Output("bank-summary", "children"),
     Output("investment-details", "children"),
     Output("summary-table", "children")],
    [Input("bank-selector", "value")]
)
def update_speedometer(selected_bank):
    fig_bar = go.Figure()
    
    # Add traces for each zone
    for zone in zones:
        fig_bar.add_trace(go.Bar(
            name=zone_labels[zone],
            x=df['Institution'],
            y=df[zone],
            marker_color=zone_colors[zone],
            text=[f"{val:.1f}%" if val > 0 else "0%" for val in df[zone]],
            textposition="inside",
            textfont=dict(size=10, color='white'),
            hovertemplate="<b>%{x}</b><br>" +
                         f"{zone_labels[zone]}: %{{y:.1f}}%<br>" +
                         "<extra></extra>"
        ))
    
    # Update bar chart layout with dark theme
    fig_bar.update_layout(
        barmode='stack',
        title={
            'text': 'Investment Distribution by Bank and Sector',
            'font': {'size': 16, 'color': '#f0f6fc'}
        },
        xaxis={
            'title': 'Bank Name',
            'tickangle': -45,
            'color': '#f0f6fc'
        },
        yaxis={
            'title': 'Investment Percentage (%)',
            'range': [0, 100],
            'color': '#f0f6fc'
        },
        height=450,
        showlegend=True,
        legend_title='Sectors',
        hovermode='x unified',
        plot_bgcolor='#161b22',
        paper_bgcolor='#161b22',
        font={'color': '#f0f6fc'}
    )
    
    # Get selected bank's data
    bank_data = df[df['Institution'] == selected_bank].iloc[0]
    risk_tier = bank_data['Climate_risk_tier']
    
    # Calculate risk score using fr weights with bank data
    risk_score = (fr.agri_w * bank_data['agri_loan_pct'] + 
                  fr.cst_w * bank_data['coastal_branches_pct'] + 
                  fr.fld_w * bank_data['flood_zone_pct'] + 
                  fr.en_w * bank_data['energy_sector_pct']) / (fr.total_w * 100)
    
    # Determine color for bank
    if risk_score < 0.33:
        bank_color = "#2ECC71"
    elif risk_score < 0.66:
        bank_color = "#F1C40F"
    else:
        bank_color = "#E74C3C"
    
    # Create bank speedometer
    fig_gauge = create_speedometer(risk_score, risk_tier, bank_color, height=350)
    
    # Create bank summary
    total_investment = bank_data[zones].sum()
    
    summary = html.Div([
        html.Strong(f" {selected_bank}", style={'fontSize': '18px', 'display': 'block', 'marginBottom': '10px', 'color': '#00c853'}),
        html.Div([
            html.Span(f"Total Investment: {total_investment:.1f}%", style={'display': 'block', 'marginBottom': '5px', 'color': '#f0f6fc'}),
            html.Span(f"Risk Tier: ", style={'display': 'inline', 'color': '#f0f6fc'}),
            html.Span(risk_tier, style={
                'color': risk_colors.get(risk_tier, 'black'),
                'fontWeight': 'bold',
                'fontSize': '16px',
                'display': 'inline'
            }),
            html.Br(),
            html.Span(f"Risk Score: {(risk_score*100):.1f}%", 
                     style={'display': 'inline-block', 'marginTop': '5px', 'color': '#f0f6fc'})
        ])
    ])
    
    # Create investment details
    details = html.Div([
        html.Strong("Investment Breakdown:", style={'display': 'block', 'marginBottom': '10px', 'color': '#00c853'}),
        html.Div([
            html.Div([
                html.Span(f"{zone_labels[zone]}: ", style={'fontWeight': 'bold', 'color': '#f0f6fc'}),
                html.Span(f"{bank_data[zone]:.1f}%", style={'color': '#f0f6fc'}),
                html.Span(" ▲" if bank_data[zone] > df[zone].mean() else " ▼",
                         style={'color': '#2ECC71' if bank_data[zone] > df[zone].mean() else '#E74C3C'})
            ], style={'marginBottom': '5px', 'padding': '5px', 'borderBottom': '1px solid #00c85333'}) 
            for zone in zones
        ])
    ])
    
    # ===== CREATE SUMMARY TABLE WITH DARK THEME =====
    table_data = df.copy()
    table_data['Total_Investment'] = table_data[zones].sum(axis=1)
    table_data['Risk_Score'] = table_data[zones].mean(axis=1) / 100
    
    # Create table using Dash HTML components with dark theme
    table_header = html.Tr([
        html.Th('Bank', style={'padding': '12px', 'backgroundColor': '#00c853', 'color': '#0d1117', 'fontWeight': 'bold'}),
        html.Th('Agriculture %', style={'padding': '12px', 'backgroundColor': '#00c853', 'color': '#0d1117', 'fontWeight': 'bold'}),
        html.Th('Coastal %', style={'padding': '12px', 'backgroundColor': '#00c853', 'color': '#0d1117', 'fontWeight': 'bold'}),
        html.Th('Flood %', style={'padding': '12px', 'backgroundColor': '#00c853', 'color': '#0d1117', 'fontWeight': 'bold'}),
        html.Th('Energy %', style={'padding': '12px', 'backgroundColor': '#00c853', 'color': '#0d1117', 'fontWeight': 'bold'}),
        html.Th('Total %', style={'padding': '12px', 'backgroundColor': '#00c853', 'color': '#0d1117', 'fontWeight': 'bold'}),
        html.Th('Risk Tier', style={'padding': '12px', 'backgroundColor': '#00c853', 'color': '#0d1117', 'fontWeight': 'bold'})
    ])
    
    table_rows = []
    for _, row in table_data.iterrows():
        # Highlight selected bank with green glow
        bg_color = '#00c85322' if row['Institution'] == selected_bank else '#161b22'
        text_color = '#f0f6fc'
        
        table_rows.append(html.Tr([
            html.Td(row['Institution'], style={
                'padding': '10px', 
                'fontWeight': 'bold' if row['Institution'] == selected_bank else 'normal', 
                'backgroundColor': bg_color,
                'color': text_color,
                'borderBottom': '1px solid #00c85333'
            }),
            html.Td(f"{row['agri_loan_pct']:.1f}%", style={
                'padding': '10px', 
                'backgroundColor': bg_color,
                'color': text_color,
                'borderBottom': '1px solid #00c85333'
            }),
            html.Td(f"{row['coastal_branches_pct']:.1f}%", style={
                'padding': '10px', 
                'backgroundColor': bg_color,
                'color': text_color,
                'borderBottom': '1px solid #00c85333'
            }),
            html.Td(f"{row['flood_zone_pct']:.1f}%", style={
                'padding': '10px', 
                'backgroundColor': bg_color,
                'color': text_color,
                'borderBottom': '1px solid #00c85333'
            }),
            html.Td(f"{row['energy_sector_pct']:.1f}%", style={
                'padding': '10px', 
                'backgroundColor': bg_color,
                'color': text_color,
                'borderBottom': '1px solid #00c85333'
            }),
            html.Td(f"{row['Total_Investment']:.1f}%", style={
                'padding': '10px', 
                'fontWeight': 'bold', 
                'backgroundColor': bg_color,
                'color': '#00c853',
                'borderBottom': '1px solid #00c85333'
            }),
            html.Td(row['Climate_risk_tier'], style={
                'padding': '10px',
                'fontWeight': 'bold',
                'color': risk_colors.get(row['Climate_risk_tier'], 'white'),
                'backgroundColor': bg_color,
                'borderBottom': '1px solid #00c85333'
            })
        ]))
    
    table = html.Table([table_header] + table_rows, style={
        'width': '100%',
        'borderCollapse': 'collapse',
        'fontFamily': 'Segoe UI, Arial, sans-serif',
        'backgroundColor': '#161b22',
        'borderRadius': '8px',
        'overflow': 'hidden'
    })
    
    return fig_bar, fig_gauge, summary, details, table

if __name__ == '__main__':
    PORT = int(os.environ.get("PORT", 8051))
    app.run(host='0.0.0.0',debug=False, port=PORT, use_reloader=False)
```
